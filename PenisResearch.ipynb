## Forecasting and Market Sizing

### Subtask:
Estimate potential market sizes for products or services by understanding the distribution of measurements within different countries or regions.


**Reasoning**:
To define the size categories as instructed, I will extract the 25th, 50th (median), and 75th percentile values for 'Erect Length (cm)' and 'Erect Circumference (cm)' from the `descriptive_stats` DataFrame. These quartiles will serve as the thresholds for 'Small', 'Medium', and 'Large' size categories, which will then be stored for later use.



````python
erect_length_q1 = descriptive_stats.loc['25%', 'Erect Length (cm)']
erect_length_q2 = descriptive_stats.loc['50%', 'Erect Length (cm)'] # Median
erect_length_q3 = descriptive_stats.loc['75%', 'Erect Length (cm)']

erect_circumference_q1 = descriptive_stats.loc['25%', 'Erect Circumference (cm)']
erect_circumference_q2 = descriptive_stats.loc['50%', 'Erect Circumference (cm)'] # Median
erect_circumference_q3 = descriptive_stats.loc['75%', 'Erect Circumference (cm)']

size_thresholds = {
    'Erect Length (cm)': {
        'Q1': erect_length_q1,
        'Q2': erect_length_q2,
        'Q3': erect_length_q3
    },
    'Erect Circumference (cm)': {
        'Q1': erect_circumference_q1,
        'Q2': erect_circumference_q2,
        'Q3': erect_circumference_q3
    }
}

print("Defined size category thresholds based on global quartiles:")
for measurement, thresholds in size_thresholds.items():
    print(f"  {measurement}:")
    print(f"    Small (<={thresholds['Q1']:.2f} cm)")
    print(f"    Medium (> {thresholds['Q1']:.2f} cm and <= {thresholds['Q3']:.2f} cm)")
    print(f"    Large (> {thresholds['Q3']:.2f} cm)")
````

**Reasoning**:
Now that the size thresholds are defined, I will apply these categories to the `regional_means` DataFrame to classify each region based on its average 'Erect Length (cm)' and 'Erect Circumference (cm)'. This will involve creating new columns to store these classifications, addressing instruction #2.



````python
def get_size_category(value, thresholds):
    if value <= thresholds['Q1']:
        return 'Small'
    elif value > thresholds['Q1'] and value <= thresholds['Q3']:
        return 'Medium'
    else:
        return 'Large'

# Apply sizing to regional_means for Erect Length
regional_means['Erect Length Category'] = regional_means['Erect Length (cm)'].apply(
    lambda x: get_size_category(x, size_thresholds['Erect Length (cm)'])
)

# Apply sizing to regional_means for Erect Circumference
regional_means['Erect Circumference Category'] = regional_means['Erect Circumference (cm)'].apply(
    lambda x: get_size_category(x, size_thresholds['Erect Circumference (cm)'])
)

print("Regional means with size categories:")
print(regional_means)
````

````python
print("\nCounts of Regions per Erect Length Category:")
print(regional_means['Erect Length Category'].value_counts())

print("\nCounts of Regions per Erect Circumference Category:")
print(regional_means['Erect Circumference Category'].value_counts())
````
