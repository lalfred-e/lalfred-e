### Region with the Most Viral Content

Next, let's find out which `location` (region) tends to generate the most viral content by looking at the average viral coefficient per region.

```python
# Group by location and calculate the mean viral coefficient
region_viral_performance = df.groupby('location')['viral_coefficient'].mean().reset_index()

# Sort to find the most viral regions
region_viral_performance = region_viral_performance.sort_values(by='viral_coefficient', ascending=False)

print("Average Viral Coefficient by Region (Most Viral First):\n")
display(region_viral_performance)

# Visualize the regional viral performance
plt.figure(figsize=(12, 6))
sns.barplot(x='location', y='viral_coefficient', data=region_viral_performance, palette='viridis')
plt.title('Average Viral Coefficient by Region')
plt.xlabel('Region')
plt.ylabel('Average Viral Coefficient')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Media Type Engagement (Comments, Shares, Views, Likes)

Finally, let's analyze which media types are most effective at generating different forms of engagement: comments, shares, views, and likes.

```python
# Group by media_type and calculate the mean of comments, shares, views, and likes
media_type_engagement = df.groupby('media_type')[['comments', 'shares', 'views', 'likes']].mean().reset_index()

print("Average Engagement Metrics by Media Type:\n")
display(media_type_engagement.sort_values(by=['likes', 'comments', 'shares', 'views'], ascending=False))

# Visualize the engagement metrics for each media type
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Average Engagement Metrics by Media Type', fontsize=16)

sns.barplot(x='media_type', y='comments', data=media_type_engagement.sort_values('comments', ascending=False), ax=axes[0, 0], palette='Blues')
axes[0, 0].set_title('Average Comments by Media Type')
axes[0, 0].set_xlabel('')
axes[0, 0].set_ylabel('Average Comments')

sns.barplot(x='media_type', y='shares', data=media_type_engagement.sort_values('shares', ascending=False), ax=axes[0, 1], palette='Greens')
axes[0, 1].set_title('Average Shares by Media Type')
axes[0, 1].set_xlabel('')
axes[0, 1].set_ylabel('Average Shares')

sns.barplot(x='media_type', y='views', data=media_type_engagement.sort_values('views', ascending=False), ax=axes[1, 0], palette='Reds')
axes[1, 0].set_title('Average Views by Media Type')
axes[1, 0].set_xlabel('')
axes[1, 0].set_ylabel('Average Views')

sns.barplot(x='media_type', y='likes', data=media_type_engagement.sort_values('likes', ascending=False), ax=axes[1, 1], palette='Purples')
axes[1, 1].set_title('Average Likes by Media Type')
axes[1, 1].set_xlabel('')
axes[1, 1].set_ylabel('Average Likes')

for ax in axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.show()
```
